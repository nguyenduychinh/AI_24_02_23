```python
from skfuzzy.control.controlsystem import ControlSystemSimulation
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
temp= ctrl.Antecedent(np.arange(20.4,41.2,0.1),'Temp')
wind = ctrl.Antecedent(np.arange(0.7,3.1,0.1),'Wind')
flow = ctrl.Consequent(np.arange(0.2,9.1,0.1),'Flow')

temp['ratthap']=fuzz.trimf(temp.universe,[20.4,20.4,26])
temp['thap']=fuzz.trimf(temp.universe,[20.4,26,30])
temp['binhthuong']=fuzz.trimf(temp.universe,[26,30,37])
temp['cao']=fuzz.trimf(temp.universe,[30,37.2,41.2])
temp['ratcao']=fuzz.trimf(temp.universe,[37,41.2,41.2])
wind['ratnhe']=fuzz.trimf(wind.universe,[0.7,0.7,1.3])
wind['nhe']=fuzz.trimf(wind.universe,[0.7,1.3,2])
wind['binhthuong']=fuzz.trimf(wind.universe,[1.3,2,2.6])
wind['manh']=fuzz.trimf(wind.universe,[2,2.6,3.1])
wind['ratmanh']=fuzz.trimf(wind.universe,[2.6,3.1,3.1])
flow['VL']=fuzz.trimf(flow.universe,[0.2,0.2,2.8])
flow['L']=fuzz.trimf(flow.universe,[0.2,2.8,5])
flow['N']=fuzz.trimf(flow.universe,[2.8,5,7])
flow['H']=fuzz.trimf(flow.universe,[5,7,9.1])
flow['VH']=fuzz.trimf(flow.universe,[7,9.1,9.1])

luat1 = ctrl.Rule(temp['ratthap'] & wind['ratnhe'],flow['VL'])
luat2 = ctrl.Rule(temp['ratthap'] & wind['nhe'],flow['VL'])
luat3 = ctrl.Rule(temp['ratthap'] & wind['binhthuong'],flow['VL'])
luat4 = ctrl.Rule(temp['ratthap'] & wind['manh'],flow['L'])
luat5 = ctrl.Rule(temp['ratthap'] & wind['ratmanh'],flow['L'])
luat6 = ctrl.Rule(temp['thap'] & wind['ratnhe'],flow['VL'])
luat7 = ctrl.Rule(temp['thap'] & wind['nhe'],flow['VL'])
luat8 = ctrl.Rule(temp['thap'] & wind['binhthuong'],flow['VL'])
luat9 = ctrl.Rule(temp['thap'] & wind['manh'],flow['L'])
luat10 = ctrl.Rule(temp['thap'] & wind['ratmanh'],flow['N'])
luat11 = ctrl.Rule(temp['binhthuong'] & wind['ratnhe'],flow['L'])
luat12 = ctrl.Rule(temp['binhthuong'] & wind['nhe'],flow['L'])
luat13 = ctrl.Rule(temp['binhthuong'] & wind['binhthuong'],flow['L'])
luat14 = ctrl.Rule(temp['binhthuong'] & wind['manh'],flow['N'])
luat15 = ctrl.Rule(temp['binhthuong'] & wind['ratmanh'],flow['H'])
luat16 = ctrl.Rule(temp['cao'] & wind['ratnhe'],flow['L'])
luat17 = ctrl.Rule(temp['cao'] & wind['nhe'],flow['N'])
luat18 = ctrl.Rule(temp['cao'] & wind['binhthuong'],flow['N'])
luat19 = ctrl.Rule(temp['cao'] & wind['manh'],flow['H'])
luat20 = ctrl.Rule(temp['cao'] & wind['ratmanh'],flow['VH'])
luat21 = ctrl.Rule(temp['ratcao'] & wind['ratnhe'],flow['N'])
luat22 = ctrl.Rule(temp['ratcao'] & wind['nhe'],flow['N'])
luat23 = ctrl.Rule(temp['ratcao'] & wind['binhthuong'],flow['H'])
luat24 = ctrl.Rule(temp['ratcao'] & wind['manh'],flow['VH'])
luat25 = ctrl.Rule(temp['ratcao'] & wind['ratmanh'],flow['VH'])

tipping_ctrl = ctrl.ControlSystem([luat1,luat2,luat3,luat4,luat5,luat6,luat7,luat8,luat9,luat10,luat11,luat12,luat13,luat14,luat15,luat16,luat17,luat18,luat19,luat20,luat21,luat22,luat23,luat24,luat25])
tipping = ControlSystemSimulation(tipping_ctrl)
tipping.input['Temp']=38
tipping.input['Wind']=3
tipping.compute()
print(tipping.output['Flow'])
flow.view(sim = tipping)






```
